## Show some initial data

```python
salary = pd.read_csv('/kaggle/input/salary/Salary.csv')
salary.head()
```

## Selecting the Features

```python
X = salary['YearsExperience'].values
y = salary['Salary'].values
```

## X (Independent Data)

```python
print(X)
```

## y (Dependent Data)

```python
print(y)
```

```python
X = X.reshape(-1, 1)
print(X)
```

## Keep track of our data
- X feature = years of experience
- y = salary in $
- We are going to manipulate the data using train_test_split()

```python
var1, var2 = 2, 3
print(var1, var2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8, test_size=.2, random_state=100)
```

```python
print(f'X_training shape [{X_train}]')
print(f'X_test shape [{X_test}]')
print(f'y_training shape [{y_train}]')
print(f'y_test shape [{y_test}]')
```

```python
plt.scatter(X_train, y_train, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary in $')
plt.title('Training Data')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_predict = lr.predict(X_test)
```

```python
print(f'Trainining Acc {round(lr.score(X_train, y_train) * 100, 2)}%')
print(f'Testing Acc {round(lr.score(X_test, y_test) * 100, 2)}%')
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_test, y_predict)
plt.xlabel('Years of Experience')
plt.ylabel('Salary in $')
plt.title('Training Data')
plt.show()
```
